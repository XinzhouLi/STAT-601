Elements selection(condition)

```python
matrix=np.random.randint(1, 10,(5,5))
matrix
```

```python
new_matrix=matrix[matrix>5]
new_matrix
```

```python
new_matrix=matrix[matrix%2==0]
new_matrix
```

```python
#replace odd elements by -3, replace even elements by 0
matrix[matrix %2==1]=-3
matrix[matrix%2==0] =0

matrix
```
