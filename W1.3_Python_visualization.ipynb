# Seaborn
https://seaborn.pydata.org/

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#scikit-learn
from sklearn.datasets import load_breast_cancer
cancer =load_breast_cancer()
type(cancer)
print(cancer['data'].shape)
print(cancer['target'].shape)
```

```python
cancer
```

```python
#np.c_ Translates slice objects to concatenation along the second axis.

df_cancer=pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns =np.append(cancer['feature_names'],['target']))
```

```python
df_cancer.head(10)
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x='mean area', y='mean smoothness',data=df_cancer)
```

```python
sns.scatterplot(x='mean area', y='mean smoothness', hue ='target',data=df_cancer)
```

```python
plt.figure(figsize=(4,4))
sns.countplot(df_cancer['target'],label='Count')
```

# Seaborn pairplot, displot and heatmaps

```python
sns.pairplot(df_cancer, hue='target',vars=['mean radius','mean texture','mean perimeter','mean smoothness'])
```

```python
plt.figure(figsize=(30,30))
sns.heatmap(df_cancer.corr(),annot=True)
```

## distplot (distribution plot) combines matplotlib histogram function with kdeplot() (kernel density estimate. KDE is used to plot the probability density of a continuous variable

```python
sns.distplot(df_cancer['mean area'], bins= 30, color='g')
```

Plot two displot for each target class, 0 and 1

```python
class_0_df=df_cancer[df_cancer['target']==0]
class_1_df=df_cancer[df_cancer['target']==1]
```

```python
class_1_df
```

```python
sns.distplot(class_0_df['mean area'], bins= 30, color='g')
sns.distplot(class_1_df['mean area'], bins= 30, color='b')
```
